```python
import os
import numpy as np
import pandas as pd
from random import randint
pd.options.display.max_rows = None



def fum1(maxPower=7500000):
    return randint(0,maxPower)
def fum2(P, reduce=0.75):
    return P*reduce
def fum3(P, offset=1000000):
    if(P>offset):
        return P-offset
    else:
        return 0

def fu_data(data0, failNum=2, minFaulty=3, maxFaulty=7, minBroken=3, maxBroken=5, fum=0, maxPower=7500000, reduce=0.75, offset=1000000):
    n=failNum
    m=0
    s=8400/n
        
    while n>0:
        n=n-1
        faultyStart=randint(300+m*s,8400-s*n)
        brokenStart=randint(minFaulty*24,maxFaulty*24)+faultyStart
        brokenEnd=randint(minBroken*24,maxBroken*24)+brokenStart
        m=m+1
        i=0
        if (fum==0):
            method=randint(1,3)
        else:
            method=fum
        for p in data0['P']:
            i=i+1
            if ((i>faultyStart)&(i<brokenStart)):
                if method==1:
                    data0.iloc[i, data0.columns.get_loc('P')] = fum1(maxPower)
                elif method==2:
                    data0.iloc[i, data0.columns.get_loc('P')] = fum2(p,reduce)
                elif method==3:
                    data0.iloc[i, data0.columns.get_loc('P')] = fum3(p, offset)
            elif ((i>brokenStart)&(i<brokenEnd)):
                data0.iloc[i, data0.columns.get_loc('P')] = 0
        return data0
    
data=pd.read_csv('data/power_prod.csv', names = ['time', 'P'])
data_fu=fu_data(data)
```
